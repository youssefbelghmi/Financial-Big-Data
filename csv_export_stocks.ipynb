```python
import os
import pandas as pd

# Chemin du dossier principal
main_folder = "/Users/hamzamo/Downloads/Financial Big Data/SP&500/60seconds"

# Parcourir chaque dossier de stock
for stock_folder in os.listdir(main_folder):
    stock_path = os.path.join(main_folder, stock_folder)
    if os.path.isdir(stock_path):
        # Liste pour accumuler les DataFrames
        stock_dataframes = []
        
        # Parcourir tous les fichiers parquet dans le dossier du stock
        for file in os.listdir(stock_path):
            if file.endswith(".parquet"):
                file_path = os.path.join(stock_path, file)
                # Lire le fichier Parquet et l'ajouter à la liste
                stock_dataframes.append(pd.read_parquet(file_path))
        
        # Concaténer tous les DataFrames en un seul
        if stock_dataframes:
            combined_df = pd.concat(stock_dataframes, ignore_index=True)

            #Sorting the DataFrame by datetime
            combined_df['index'] = pd.to_datetime(combined_df['index'])
            combined_df = combined_df.sort_values(by='index')
            
            # Exporter le DataFrame en CSV
            csv_file_path = os.path.join(main_folder, f"{stock_folder}.csv")
            combined_df.to_csv(csv_file_path, index=False)
            
            print(f"DataFrame pour le stock {stock_folder} exporté dans {csv_file_path}")
```
